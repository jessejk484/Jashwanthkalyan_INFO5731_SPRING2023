# Question 1

(20 points). Write a Python program to generate 12 random numbers between 1 and 100. Sort the list of numbers in ascending order and print it out. Sort the list of numbers in descending order and print it out. Calculate the average for the group and print it out.

```python
# Your code here
import random
from statistics import mean

rand_num = []
print("Random integers between 1 and 100: ")
for i in range(10):
    y = random.randrange(100)
    rand_num.append(y)
rand_num.sort()
print("List of numbers in ascending order is: ", rand_num)
rand_num.sort(reverse=True)
print("List of numbers in descending order is: ", rand_num)
avg=mean(rand_num)
print("Average of random numbers is: ", avg)

```

# Question 3 

(15 points). The formula that describes the volume V of a sphere with radius r is the following:

$ V=\frac{4}{3}\ast\ \pi\ast\ r^3 $

Write a program to calculate the value of V when r is in the range of 1–10. Output the result in the following format:

	r	V
	1	…
	2	…
	3	…


```python
# Your code here
import math
def vol():
  for i in range(1, 11):
    yield f"{i} \t {(4/3)*(math.pi)*(i**3)}"

print("r \t V")
for v in vol():
  print(v)
```

Write a program to implement the following questions.

(1) (10 pts) Calculate the difference in the number of Occupied Housing Units from year to year and print it. The difference must be calculated for the consecutive years such as 2008-2009, 2009-2010 etc. Finally, print the values in the ascending order.

(2) (10 pts) For all the years, calculate the percentage of housing units which are vacant an occupied. Print the results in the following format:

(3) (10 pts) Calculate and print the valued and years in which the highest number of housing units were vacant and occupied. Print the results in the following format:  

                              Year	Value

(4) (10 pts) Calculate the harmonic mean of the total housing units and print it out. You can find the information about harmonic mean here: https://ncalculators.com/statistics/harmonic-mean-calculator.htm

```python
# Your code here

import pandas as pd

dataframe = pd.read_csv(r"Assignment1_denton_housing.csv")
dataframe
```

```python
ou_df  = dataframe[dataframe["title_field"] == "Occupied Housing Units"]
ou_df = ou_df.sort_values("year").reset_index(drop=True)
ou_df
```

```python
# Calculating the difference in the number of Occupied Housing Units from year to year 

print("Year period \t housing unit diff")
ud_map = {}
for row in range(1, len(ou_df)):
  different_value = ou_df.loc[row, 'value'] - ou_df.loc[row - 1, 'value']
  ud_map[f"{ou_df.loc[row - 1, 'year']} -> {ou_df.loc[row, 'year']}"] = different_value

unit_diff_map = dict(sorted(ud_map.items(), key = lambda x: x[1]))
for key, value in unit_diff_map.items():
  print(f"{key} \t {value}")
```

```python
# For all the years, calculate the percentage of housing units which are vacant and occupied.

df_vacant, df_occupied = dataframe[(dataframe["title_field"] == "Vacant Housing Units")].reset_index(drop=True), dataframe[(dataframe["title_field"] == "Occupied Housing Units")].reset_index(drop=True)
df_vacant = df_vacant.sort_values("year").reset_index(drop=True)
df_vacant
```

```python
df_occupied = df_occupied.sort_values("year").reset_index(drop=True)
df_occupied
```

```python
 
print("Year \t Vacant Housing Units \t Occupied Housing Units") 
for i in range(len(df_vacant)):
  print(f"{df_vacant.loc[i, 'year']} \t {round((df_vacant.loc[i, 'value']/(df_vacant.loc[i, 'value']+ df_occupied.loc[i, 'value']))*100, 2)}%\t \t \t {round((df_occupied.loc[i, 'value']/(df_vacant.loc[i, 'value']+ df_occupied.loc[i, 'value']))*100,2)}%")
```

```python
# Calculate and print the valued and years in which the highest number of housing units were vacant and occupied

highv_id, higho_id = df_vacant['value'].argmax(), df_occupied['value'].argmax()
print("\t \t \t Year \t Value")
print(f"Vacant Housing Units \t {df_vacant.loc[highv_id, 'year']} \t {df_vacant.loc[highv_id, 'value']}")
print(f"Occupied Housing Units \t {df_occupied.loc[higho_id, 'year']} \t {df_occupied.loc[higho_id, 'value']}")
```

```python
# Calculate the harmonic mean of the total housing units
def harmo_mean(dataframe):
  tothsing_units = dataframe[dataframe["title_field"] == "Total Housing Units"]
  tothsing_units_values = list(tothsing_units.value)
  sumth_unit = 0
  for x in tothsing_units_values:
    sumth_unit += (1/x)
  return len(tothsing_units_values)/sumth_unit
print(f"Harmonic Mean of the total housing units: {harmo_mean(dataframe)}")
```
